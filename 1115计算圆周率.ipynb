```python
import numpy as np
import random
import dask
from dask.distributed import Client

client = Client(n_workers=4)
```

```python
client
```

```python
%%time
n = 10**7
x = np.random.random(n)
y = np.random.random(n)
@dask.delayed
def com(nl,xl,yl):
    n2 = 0
    for k in range(nl):
        if (xl[k]**2+yl[k]**2 <= 1):
            n2 = n2 + 1
    return n2
n2 = com(n,x,y)
pi = (4*n2)/n
pi.compute()
```
